```python
import pandas as pd

df = pd.read_csv('place_info.csv')

df_store = df.drop_duplicates()
print(df_store)
```

```python
import os
import requests
import sqlite3
from dotenv import load_dotenv

# 카카오 API 설정
load_dotenv()
API = os.getenv('KAKAO_API')
headers = {"Authorization": f"KakaoAK {API}"} #REST API 키(유효한 키)
url = "https://dapi.kakao.com/v2/local/search/address.json" #요청할 url 주소

rest_info_db = []
rest_info_failed = []
# 가게 별 순회
for _, row in df_store.iterrows():
    # print(row['가게 주소'])
    query = {'query': row['가게 주소']} #입력할 주소
    try:
        result = requests.get(url,
                            headers=headers,
                            params=query).json() #카카오 API 요청
        
        if not result['documents']:
            print(f'{row["가게 이름"]} 실패')
            rest_info_failed.append(row)
            continue

        rest_info_db.append({
            'name' : row['가게 이름'],
            'address' : row['가게 주소'],
            'lat' : result["documents"][0]['y'],
            'lng' : result["documents"][0]['x'],
            'category' : row['카테고리'],
        })

        print(f'{row['가게 이름']} 완')
    except:
        print(f'{row["가게 이름"]} 실패')
        rest_info_failed.append(row)


print('좌표계 변환 완\n실패목록')
print(rest_info_failed)

# df_cat = df_store[['카테고리']].drop_duplicates()
# print(df_cat)
```

```python
coord = {
    'FAFAS' : (37.296738755265174, 126.9694594726788),
    '갓덴스시 스타필드 수원점' : (37.28727615186458, 126.99106672470307)
}

for row in rest_info_failed:
    print(row)
    rest_info_db.append({
        'name' : row['가게 이름'],
        'address' : row['가게 주소'],
        'lat' : coord[row['가게 이름']][0],
        'lng' : coord[row['가게 이름']][1],
        'category' : row['카테고리'],
    })
```

```python
# SQLITe
conn = sqlite3.connect('app.sqlite3')

df_rest_info = pd.DataFrame(rest_info_db).drop_duplicates()
df_rest_info.index.name = 'id'
df_rest_info['created_at'] = '2026-06-20'
df_rest_info['updated_at'] = '2026-06-20'

df_rest_info.to_sql(
    'restaurants',
    conn,
    if_exists='replace',
)

conn.close
```

```python
df_rest_info.to_csv('place_info_coord.csv')
```
